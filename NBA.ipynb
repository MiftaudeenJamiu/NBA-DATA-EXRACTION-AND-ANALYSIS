```python
# Initialize the DataFrame
df_cols = ['Year', 'Season_type']  # To add Year and Seasons Column
df = pd.DataFrame(columns=df_cols)

season_types = ['Regular%20Season', 'Playoffs']
years = ['2014-15', '2015-16', '2016-17', '2017-18', '2018-19', '2019-20', '2020-21', '2021-22', '2022-23', '2023-24']

begin_loop = time.time()

for y in years:
    for s in season_types:
        api_url = f'https://stats.nba.com/stats/leagueLeaders?LeagueID=00&PerMode=Totals&Scope=S&Season={y}&SeasonType={s}&StatCategory=PTS'
        r = requests.get(url=api_url).json()

        # Use headers from API response
        if 'resultSet' in r and 'rowSet' in r['resultSet']:
            table_headers = r['resultSet']['headers']  # Get headers dynamically
            temp_df1 = pd.DataFrame(r['resultSet']['rowSet'], columns=table_headers)
            
            temp_df2 = pd.DataFrame({
                'Year': [y for _ in range(len(temp_df1))],
                'Season_type': [s for _ in range(len(temp_df1))]
            })
            
            temp_df3 = pd.concat([temp_df2, temp_df1], axis=1)
            df = pd.concat([df, temp_df3], axis=0)

            print(f'Finished scraping for the {y} {s}.')
        else:
            print(f"No data found for {y} {s}")

        lag = np.random.uniform(low=5, high=40)
        print(f'...waiting {round(lag, 1)} seconds')
        time.sleep(lag)

print(f'Process completed! Total run time: {round((time.time() - begin_loop) / 60)} minutes')

df.to_csv('nba_data.csv', index=False, encoding='utf-8') #saving our file in csv format
```
